```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
pic = Image.open('imgpuppy.jpg')
type(pic)
```

```python
pic_arr = np.asarray(pic)
type(pic_arr)
```

```python
print(pic_arr.shape)
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
pic_red[:,:,1:] = 0 #setting the other channels to zero
plt.imshow(pic_red)
```

```python
pic_blue = pic_arr.copy()
pic_blue[:,:,0] = 0 #setting the other channels to zero
pic_blue[: , : , 1] = 0 #setting the other channels to zero
plt.imshow(pic_blue)
```

```python
pic_green = pic_arr.copy()
pic_green[:,:,0] = 0 #setting the other channels to zero
pic_green[: , : , 2] = 0 #setting the other channels to zero
plt.imshow(pic_green)
```
