```python
import cv2
img1 = cv2.imread('puppyimg2.jpg')
img2 = cv2.imread('imgpuppy.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

```python
#BLENDING IMAGES OF THE SAME SIZE
img1 = cv2.resize(img1 , (1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1 , alpha=0.9,src2=img2,beta = 0.1,gamma=0)
plt.imshow(blended)
```

```python
#OVERLAY a small image on top of a larger image.
#numpy reassignment operation
img2 = cv2.resize(img2,(600,600))
plt.imshow(img2)
```

```python
large_img = img1
small_image = img2
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset +small_image.shape[1]
y_end = y_offset + small_image.shape[0]
```

```python
large_img[y_offset : y_end , x_offset:x_end] = small_image
plt.imshow(large_img)
```
